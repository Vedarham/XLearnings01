```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("yasserh/titanic-dataset")
print(os.listdir(path))
```

```python
df = pd.read_csv(os.path.join(path,'Titanic-Dataset.csv'))
```

```python
df.drop(columns = ['PassengerId','Name','Cabin','Ticket'], inplace =True)
```

```python
df.head()
```

```python
X_train , X_test, Y_train, Y_test = train_test_split(df.drop(columns = ['Survived']), df['Survived'], test_size = 0.2, random_state = 42)
```

```python
X_train.head()
```

```python
si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy = 'most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
ohe_gender = OneHotEncoder(sparse_output = False, dtype = np.int32, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output = False, dtype = np.int32,  handle_unknown='ignore')

X_train_gender = ohe_gender.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_gender = ohe_gender.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_train_gender[:2]
```

```python
X_train_embarked[:2]
```

```python
X_train_left = X_train.drop(columns=['Sex','Age','Embarked'])
X_test_left = X_test.drop(columns = ['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_left, X_train_age,X_train_gender,X_train_embarked), axis=1 )
X_test_transformed =  np.concatenate((X_test_left, X_test_age,X_test_gender,X_test_embarked), axis=1 )
```

```python
X_train_transformed[:2]
```

```python
X_train_transformed.shape
```

```python
X_test_transformed.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed, Y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, y_pred)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x_train , x_test, y_train , y_test = train_test_split(df.drop(columns = ['Survived']), df['Survived'], test_size=0.2, random_state = 42)
```

```python
trf1 = ColumnTransformer([
    ('si_age' , SimpleImputer(), [2]),
    ('si_embarked', SimpleImputer(strategy = 'most_frequent'),[6])
],remainder = 'passthrough')
```

```python
# trf2 = ColumnTransformer([
#     ('ohe_sex', OneHotEncoder(handle_unknown='ignore', sparse_output=False),[1] ),
#     ('ohe_embarked', OneHotEncoder(handle_unknown='ignore', sparse_output=False),[6])
# ],remainder='passthrough')

trf2 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(handle_unknown='ignore', sparse_output=False),[1,6])
],remainder='passthrough')
```

```python
trf3 = ColumnTransformer([
    ('scaler', MinMaxScaler(),slice(0,10))
])
```

```python
from sklearn.feature_selection import SelectKBest, chi2
trf4 = SelectKBest(score_func = chi2, k=8)
```

```python
trf5 = DecisionTreeClassifier()
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5),
])
```

```python
# pipe = make_pipeline(tnf1,tnf2,tnf3,tnf4,tnf5)
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.named_steps
```

```python
# Checking Parameters and their lookup of output is easier , thereby easy backtracking
pipe.named_steps['trf1'].transformers_[0][1].statistics_
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
