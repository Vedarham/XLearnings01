```python
import pandas as pd
import os
import numpy as np
```

```python
import kagglehub

# Download latest version
path1 = kagglehub.dataset_download("rohitgrewal/airlines-flights-data")

print("Path to dataset files:", path1)
print(os.listdir(path1))
flights  =pd.read_csv(os.path.join(path1, "airlines_flights_data.csv"))
```

```python
import kagglehub

# Download latest version
path2 = kagglehub.dataset_download("brendan45774/test-file")

print(os.listdir(path2))

titanic =pd.read_csv(os.path.join(path2, "tested.csv"))
```

```python
import kagglehub

# Download latest version
path3 = kagglehub.dataset_download("himanshunakrani/iris-dataset")

print("Path to dataset files:", path3)
print(os.listdir(path3))
iris =pd.read_csv(os.path.join(path3, "iris.csv"))
```

```python
import seaborn as sns
```

**Bivariate Analysis**

This is Numerical to Nummerical

```python
sns.scatterplot(y = flights['duration'],x =flights['price'])
```

**Multivariate Analysis**

```python
sample_flights = flights.sample(300)
sns.scatterplot(y = sample_flights['duration'],x =sample_flights['price'], hue = sample_flights['airline'],size=sample_flights['stops'])
```

**For Numerical - Categorical Representation**

```python
titanic.head()
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'])
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'], hue = titanic['Sex'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

**HeatMap**

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

**Better measure is percentages**

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

**ClusterMap**

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
